```python
df=df.drop_duplicates()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df
```

```python
df.describe() 
```

```python
df["Avg_Score"] = (
    df["Math_Score"] +
    df["Science_Score"] +
    df["English_Score"] +
    df["History_Score"]
) / 4

df["Avg_Score"]
```

```python
def attendance(x):
    if x >= 90:
        return "Excellent"
    elif x >= 75:
        return "Good"
    else:
        return "Poor"

df["Attendance_Category"] = df["Attendance_%"].apply(attendance)
df["Attendance_Category"]
df
```

```python
df["Pass/Fail"] = df["Avg_Score"].apply(
    lambda x: "Pass" if x >= 50 else "Fail"
)
df
```

```python
df.to_csv("D:\\edge downlode\\cleaned_education_dataset.csv")
```

```python
from sqlalchemy import create_engine
engine = create_engine(
    "mysql+pymysql://root:@localhost:3306/education_db"
)
df.to_sql(
    name="education_dataset",
    con=engine,
    if_exists="replace",
    index=False
)
```
